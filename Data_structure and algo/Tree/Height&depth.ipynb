```python
#structure of binary tree
class Node:
  def __init__(self,key):
    self.data = key
    self.left = None
    self.right = None
```

```python
#find the depth of given node
def findpath(root,key):
  if (root == None):
    return -1
  #initialize distance with -1
  dist = -1
  #check if the key is the current node
  if (root.data==key) :
    return dist + 1
  #print("left:",dist)
  dist = findpath(root.left, key)
  if dist >= 0:
    return dist + 1
  #print("right:",dist)
  dist = findpath(root.right, key)
  if dist >= 0:
    return dist+1
  return dist
```

```python
#helper function for height
def findheightunit(root,x):
  global height
  if (root == None):
    return -1
  leftheight = findheightunit(root.left,x)
  rightheight = findheightunit(root.right,x)
  ans = max(leftheight,rightheight) + 1
  if (root.data == x):
    height = ans
  return ans
```
