```python
#structure of a node
class Node:
  def __init__(self, key):
    self.key = key
    self.child = []
```

```python
#function to create a new node
def new_node(key):
  temp = Node(key)
  return temp
```

```python
#function for returning depth
def depth_tree(pr):
  if pr is None:
    return 0
  max_depth = 0
  for ch in pr.child:
    max_depth = max(max_depth,depth_tree(ch))
  return max_depth+1
```

```python
if __name__ == '__main__':
  root = new_node('A')
  root.child.extend([new_node('B'),new_node('F'),new_node('D'),new_node('E')])
  root.child[0].child.extend([new_node('K'),new_node('J')])
  root.child[2].child.append(new_node('G'))
  root.child[3].child.extend([new_node('C'),new_node('H'),new_node('I')])
  root.child[0].child[0].child.extend([new_node('N'),new_node('M')])
  root.child[3].child[2].child.append(new_node('L'))
  print("depth of the tree:",depth_tree(root))


```
