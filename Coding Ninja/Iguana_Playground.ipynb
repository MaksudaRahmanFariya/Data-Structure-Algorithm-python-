```python
def max_satisfaction(N, F, S, satisfaction):
  dp = [[-float('inf')] * (S + 1) for _ in range(F + 1)]
  #dp = [[0] * (S + 1) for _ in range(F + 1)]
  for i in range(N):
    fi, si = satisfaction[i]
    new_dp = [row[:] for row in dp]
    for j in range(1, F + 1):
      for k in range(1, S + 1):
        new_dp[j][k] = max(dp[j][k], dp[j-1][k] + fi, dp[j][k-1] + si)
    dp = new_dp
  return max(max(row) for row in dp)

# Example usage
N, F, S = map(int, input().split())
satisfaction = [tuple(map(int, input().split())) for _ in range(N)]
print(max_satisfaction(N, F, S, satisfaction))
```
