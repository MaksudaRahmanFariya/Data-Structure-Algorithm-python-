```python
def cal_feasts(recharge_amount, current_balance):
    meal_cost = 2 #meal cost
    #number of drachmas spent
    drachmas_spent = recharge_amount - current_balance
    # number of times he has feasted
    feasts = drachmas_spent // meal_cost
    return feasts
```

```python
if __name__== '__main__':
  recharge_amount, current_balance = [int(a) for a in input().split()]
  feasts = cal_feasts(recharge_amount, current_balance)
  print(feasts)
```
