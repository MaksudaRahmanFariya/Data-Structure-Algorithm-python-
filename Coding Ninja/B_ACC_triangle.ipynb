```python
def max_triangle_height(x, y):
    l, r = 0, min(x, y + 1)
    while l < r:
        mid = (l + r + 1) // 2
        if mid - 1 <= y and (mid * (mid - 1)) // 2 <= y:
            l = mid
        else:
            r = mid - 1
    return l

```

```python
if __name__ == "__main__":
  x, y = [int(x) for x in input().split()]
  max_height = max_triangle_height(x, y)
  print(max_height)
```
