```python
def good_permutations(t, test_cases):
    results = []
    for n in test_cases:
        if n == 1:
            results.append([1])
        elif n == 2:
            results.append([1, 2])
        else:
          perm = [1] + list(range(n, 1, -1))
          results.append(perm)

    return results
```

```python
if __name__ == '__main__':
    t = int(input().strip())
    test_case = list(map(int, input().strip().split()))[:t]
    result = good_permutations(t, test_case)
    print("\n".join(" ".join(map(str, p)) for p in result))
```
